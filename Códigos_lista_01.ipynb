```python
## solução 3

#a
def prov_a():
    i = 1
    while True:
        if i <= 0:  # verifica se i é positivo
            break
        i += 1
    return i
print(prov_a())  # resultado: infinito
```

```python
#b

def prov_b():
    i = -1
    while True:
        if i >= 0:  # verifica se i é negativo
            break
        i -= 1
    return i
print(prov_b())  # resultado: -infinito
```

```python
#c
def prov_c():
    i = 2
    while True:
        if i % 2 != 0:  # verifica se i é par
            break
        i += 2
    return i
print(prov_c())  # resultado: infinito
```

```python
#d

def prov_d():
    x = 0
    while x >= 0:  # enquanto x é positivo ou igual a zero
        x += 1e-10  # aumenta o valor de x
    return x
print(prov_d())  # resultado: infinito (aproximadamente)
```
